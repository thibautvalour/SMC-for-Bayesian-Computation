```python
import numpy as np 
import matplotlib.pyplot as plt
from init import init_Y
from metrics import distance
```

```python
Y = init_Y()
results = {f'sim_{k}' : np.load(f'results_dic/results_{k}.npy', allow_pickle=True).item() for k in range(1, 11)}
```

```python
# hist of size of Xs 
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10, 5))
ax1.hist(list(map(len, results['sim_1']['Xs'][-1])), bins=50, alpha=0.5)
ax2.hist(list(map(len, results['sim_2']['Xs'][-1])), bins=50, alpha=0.5)
ax3.hist(list(map(len, results['sim_3']['Xs'][-1])), bins=50, alpha=0.5)
plt.show()
```

```python
# hist of size of max of Xs
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10, 5))
ax1.hist(list(map(max, results['sim_1']['Xs'][-1])), bins=50, alpha=0.5)
ax2.hist(list(map(max, results['sim_2']['Xs'][-1])), bins=50, alpha=0.5)
ax3.hist(list(map(max, results['sim_3']['Xs'][-1])), bins=50, alpha=0.5)
plt.show()
```

```python
# hist of size of max of Xs / sum of Xs
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10, 5))
ax1.hist(list(map(lambda L : max(L)/np.sum(L), results['sim_1']['Xs'][-1])), bins=50, alpha=0.5)
ax2.hist(list(map(lambda L : max(L)/np.sum(L), results['sim_2']['Xs'][-1])), bins=50, alpha=0.5)
ax3.hist(list(map(lambda L : max(L)/np.sum(L), results['sim_3']['Xs'][-1])), bins=50, alpha=0.5)
plt.show()
```
